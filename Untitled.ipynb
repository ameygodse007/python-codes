```python
class fruit():
    count=0
    def __init__(self,name='',size=0,color=''):
        self.name=name
        self.size=size
        self.color=color
        fruit.count+=1
```

```python
def display():
    print(fruit.count)
f1=fruit(name='banana',size=50,color='yellow')
f2=fruit(name='mango',size=3,color='orange')
print(vars(f1))
print(vars(f2))
print(display())
```
